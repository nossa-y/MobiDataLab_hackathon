```python
path = '/kaggle/input/schedules/PUJO.TMI'
data = pd.read_csv(path, sep = '|')
data = data.tail(15)
data.columns = [el[1:-1] for el in data.columns]
data.sort_values(by='DepartureTime', inplace=True)

```

```python
data['DepartureTime'] = pd.to_datetime(data['DepartureTime'], format = '%H:%M:%S')
#data.dtypes
```

```python
from datetime import datetime, timedelta
```

```python
creux = []
taxis_sch = []
for i in range(1, len(data)):
    diff = data.iloc[i]['DepartureTime'] - data.iloc[i - 1]['DepartureTime']
    if diff > timedelta(minutes=30):
        creux.append({
            'Debut': data.iloc[i - 1]['DepartureTime'],
            'Fin': data.iloc[i]['DepartureTime'],
            'Duree': diff
        })
intervalle_10_min = timedelta(minutes=10)
liste_intervals = []

for creux in creux:
    debut = creux['Debut'] + intervalle_10_min
    while debut < creux['Fin']:
        liste_intervals.append(debut)
        debut += intervalle_10_min
liste_intervals
```
